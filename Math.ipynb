```python
import pandas as pd
import math
import datetime as dt
```

```python
with open('chat.txt', 'r') as myfile:
    data=myfile.read().replace('\n', '')
data = data.split(' ')
chatdict = {}
for item in data:
    if item not in chatdict:
        chatdict[item] = 1
    else:
        chatdict[item] += 1
for k, v in chatdict.items():
    print(str(k) + ':' + str(v))
```

```python
columns = ['Word', 'Count', 'Time', 'Weight']
df = pd.DataFrame(columns = columns)
print(df)
for k, v in chatdict:
    df = df.append({'Word': k, 'Count': v, 'Time': dt.now(), 'Weight': math.log(v)}, ignore_index=True)
print(df)
```
